```python
"""Backend supported: tensorflow.compat.v1, tensorflow, pytorch, paddle"""
import deepxde as dde
import numpy as np


def pde(r, y):
    dy_r = dde.grad.jacobian(y, r)
    r2dy_r = (r**2)*dy_r
    dr2dy_r_r = dde.grad.jacobian(r2dy_r, r)
    return (1/r**2) * dr2dy_r_r



def boundary_r(r, on_boundary):
    return on_boundary and np.isclose(r[0], 1000)


def analytic_soln(r):
    return (1 + (2/(r)))


geom = dde.geometry.Interval(2, 1000)


bc_r = dde.icbc.DirichletBC(geom, analytic_soln, boundary_r)



#bc_l = dde.icbc.RobinBC(geom, lambda X, y: 1, boundary_l)
#bc_r = dde.icbc.RobinBC(geom, lambda X, y: 1, boundary_r)


data = dde.data.PDE(geom, pde, [bc_r], 50, 500, solution=analytic_soln, num_test=200,
                   train_distribution="uniform")



layer_size = [1] + [50] * 3 + [1]
activation = "tanh"
initializer = "Glorot uniform"
net = dde.nn.FNN(layer_size, activation, initializer)



model = dde.Model(data, net)
model.compile("adam", lr=0.001, metrics=["l2 relative error"])
losshistory, train_state = model.train(iterations=1000)

dde.saveplot(losshistory, train_state, isplot=True)
```
